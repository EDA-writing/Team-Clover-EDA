| Column | Meaning |

|Country | Country

|Status | Developed or Developing status |

```python
def missing_values(df: pd.DataFrame):
    counts = df.isna().sum().to_frame(name='Count')
    percentages = (df.isna().sum()/df.count()*100).to_frame(name='%')
    missing_df = pd.concat([counts, percentages], axis=1)
    return missing_df[missing_df.Count>0].sort_values(by='Count', ascending=False)
```

```python
import pycountry_convert as pc
```

```python
def country_to_continent(country):
    try:
        country_code = pc.country_name_to_country_alpha2(country)
        continent = pc.country_alpha2_to_continent_code(country_code)
        return pc.convert_continent_code_to_continent_name(continent)
    except Exception as e:
        pass
```

```python
df['Continent'] = df.Country.map(country_to_continent)
```

```python
df[df.Continent.isna()].Country.unique()
```

```python
country_codes = {}
for country in df[df.Continent.isna()].Country.unique():    
    if '(' in country:
        idx = country.index('(')
        c = country[:idx].strip()
        code = pc.country_name_to_country_alpha2(c)     
        continent = pc.country_alpha2_to_continent_code(code)
        country_codes[country] = pc.convert_continent_code_to_continent_name(continent)   

country_codes['Republic of Korea'] = 'Asia'
country_codes['The former Yugoslav republic of Macedonia'] = 'Europe'
country_codes['Timor-Leste'] = 'Asia'
country_codes    
```

```python
df.Continent = df.Continent.fillna(df.Country.map(country_codes))
df.Continent.isna().sum()
```

```python
cont_yr_exp = df.groupby(['Continent', 'Year'])['Life expectancy'].mean().to_frame().reset_index()

sns.set(rc={"figure.figsize":(15, 8)})
sns.barplot(data=cont_yr_exp, x="Year", y="Life expectancy", hue="Continent")
```

Check effect of expenditure on healthcare on life expectancy 

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.scatter(africa_df['percentage expenditure'], africa_df['Life expectancy'])
ax1.set_title('Percentage expenditure')
ax1.set(ylabel='Life Expectancy')
ax2.scatter(africa_df['Total expenditure'], africa_df['Life expectancy'])
ax2.set_title('Total expenditure')
```

```python
yearly_df = africa_df.groupby('Year').mean().reset_index()
```

```python
fig, ax = plt.subplots()

ax1 = ax.twinx()
ax2 = ax.twinx()

ax.plot(yearly_df.Year, yearly_df['Life expectancy'])
ax.set_ylabel('Life Expectancy')

ax1.plot(yearly_df.Year, yearly_df.GDP, color='green')
ax1.set_ylabel('GDP')

ax2.plot(yearly_df.Year, yearly_df['Total expenditure'], color='red')
ax2.set_ylabel('TE')


ax.set_xlabel('Year')

plt.show()
```
