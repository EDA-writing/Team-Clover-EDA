Add new features: 
* `iso_alpha`
* `Continent`

```python
import pycountry_convert as pc

def country_to_continent(country):
    try:
        country_code = pc.country_name_to_country_alpha2(country)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        return pc.convert_continent_code_to_continent_name(continent_code)        
    except Exception as e:
        return handle_missing(country)
    
def handle_missing(country):
    if '(' in country:
        c = country[:country.index('(')].strip()
        code = pc.country_name_to_country_alpha2(c)     
        continent = pc.country_alpha2_to_continent_code(code)
        return pc.convert_continent_code_to_continent_name(continent)
    else:
            return np.NaN

```

```python
df['Continent'] = df.Country.map(country_to_continent)
```

```python
df[df.Continent.isna()].Country.unique()
```

```python
missing_country_continents = {
    'Republic of Korea': 'Asia',
    'The former Yugoslav republic of Macedonia':'Europe',
    'Timor-Leste': 'Asia'
}
```

```python
df.Continent.fillna(df.Country.map(missing_country_continents), inplace=True)
```

```python
df.head()
```

```python
df[['Year', 'Status', 'Continent']] = df[['Year', 'Status', 'Continent']].astype('category')
```

```python
def missing_values(df: pd.DataFrame):
    counts = df.isna().sum().to_frame(name='Count')
    percentages = (df.isna().sum()/df.count()*100).to_frame(name='%')
    missing_df = pd.concat([counts, percentages], axis=1)
    return missing_df[missing_df.Count>0].sort_values(by='Count', ascending=False)
```

```python
missing_values(df)
```

```python
df1 = df.copy().drop(columns='Population', axis=1).dropna()
```

### How does Africa Compare?

```python
cont_yr_exp = df1.groupby(['Continent', 'Year'], as_index=False)['Life expectancy'].mean()

sns.set(rc={"figure.figsize":(15, 8)})
sns.lineplot(data=cont_yr_exp, x="Year", y="Life expectancy", hue="Continent")
```


We check variable correlation to life expectancy in Africa vs Globally.

```python
afr_corr = df1[df1.Continent=='Africa'].corr(numeric_only=True).drop('Life expectancy')
global_corr = df1[df1.Continent!='Africa'].corr(numeric_only=True).drop('Life expectancy')

afr_global_corr = pd.DataFrame.from_dict(data={'Africa':afr_corr['Life expectancy'], 
                                               'Global':global_corr['Life expectancy']})

# rank the correlations
afr_global_corr['Africa Rank'] = afr_global_corr.Africa.abs().rank(ascending=False)
afr_global_corr['Global Rank'] = afr_global_corr.Global.abs().rank(ascending=False)

afr_global_corr.sort_values(by='Africa Rank')
```

The top correlations in Africa are: 
* HIV/AIDS
* Income composition of resources
* Adult Mortality
* Schooling
* BMI

#### Notable differences:

* HIV/AIDS  
    `HIV/AIDS` ranks *first* in Africa compared to *tenth* in the rest of the world. This tells that `HIV/AIDS` is quite a menace in Africa as opposed to everywhere else.

* BMI and Thinness  
    Fifth in Africa while 9th elsewhere, `BMI` seems to be particularly important in the African context. `thinness 1-19 years` and `thinness 5-9 years` rank 7th and 8th globally, but 14th and 15th in Africa. A possible discrepancy given how high BMI ranks in Africa?

* Schooling  
    Time spent in school predicts life expectancy in Africa much less than the rest of the world, where it is the most significant variable.

* Vaccination  
    `Diptheria`, `Hepatitis B` and `Polio` are much more significant in Africa. This might mean these diseases are pretty much eradicated elsewhere and are therefore not much of a concern there.

* Alcohol  
    Oddly, while having a high positive global correlation, alcohol consumption has a *negative* correlation in Africa.
